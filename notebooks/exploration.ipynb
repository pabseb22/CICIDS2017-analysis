```python
# Minimal example: list PCAP files and compute baseline metrics
from src.utils.pcap_reader import PcapReader, SCAPY_AVAILABLE
from src.utils.metrics import compute_basic_metrics

reader = PcapReader()
pcaps = reader.list_pcaps()
print('Found PCAPs:', pcaps)
if len(pcaps) > 0 and SCAPY_AVAILABLE:
    pkts = reader.read_pcap(pcaps[0])
    metrics = compute_basic_metrics(pkts)
    print('Metrics for', pcaps[0], metrics)
else:
    print('No PCAPs or Scapy not installed; install scapy and add PCAPs to pcap_samples/')
```
